# Selecting the Same columns

```python
terror = data[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude",
        "location","summary","attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
terror.head()
```

# Renaming the Same columns

```python
terror.rename(columns = {"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region","provstate":"State",
                    "city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location","summary":"Summary",
                    "attacktype1_txt":"Attack_Type","targtype1_txt":"Target_Type","gname":"Gang_Name","motive":"Motive",
                    "weaptype1_txt":"Weapon_Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add_Notes"
                    },inplace = True)
terror.head()
```

# Checking For Null Values

```python
terror.isnull().sum()
```

# Imputing The Null Values

```python
terror["Killed"] = terror["Killed"].fillna(0)
terror["Wounded"] = terror["Wounded"].fillna(0)
terror["Casualty"] = terror["Killed"] + terror["Wounded"]
```

```python
terror.describe()
```

The data consists of terrorist activities ranging from the Year: 1970 to 2017
Maximum number of people killed in an event were: 1570
Maximum number of people wounded in an event were: 8191
Maximum number of total casualities in an event were: 9574

# EDA
# Visualization

# Total Casualities (Killed + Wounded) in each Year

```python
yc = terror[["Year","Casualty"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind = "bar", color = "blue", figsize = (15,6))
plt.title("Year Wise Casualities", fontsize = 15)
plt.xlabel("Years", fontsize = 15)
plt.ylabel("No of Casualities", fontsize = 15)
plt.show()
```

# Year Wise Attacks

```python
attacks = terror["Year"].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind = "bar", color = "grey", figsize = (15,6), fontsize = 12)
plt.title("Attacks Timeline", fontsize = 15)
plt.xlabel("Years", fontsize = 15)
plt.ylabel("No of Attacks", fontsize = 15)
plt.show()
```

# Wounded in each Region

```python
year_wounded = terror[["Year","Wounded"]].groupby("Year").sum()
year_wounded.head()
```

# Killed in Each Year

```python
year_killed = terror[["Year","Killed"]].groupby("Year").sum()
year_killed.head()
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(2,1,1)
ax1 = fig.add_subplot(2,1,2)

# Killed
year_killed.plot(kind = 'bar', color = 'blue', figsize = (15,15), ax = ax0)
ax0.set_title("People killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("No.of People Killed")

# Wounded
year_wounded.plot(kind = 'bar', color = 'grey', figsize = (15,15), ax = ax1)
ax1.set_title("People wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("No.of People Wounded")

plt.show()
```

# Region wise Attacks

Distribution of Terrorist Attacks over Regions from 1970-2017

```python
region = pd.crosstab(terror.Year, terror.Region)
region.head()
```

```python
region.plot(kind = "area", stacked = False, alpha = 0.6, figsize = (25,15))
plt.title("Region Wise Attacks", fontsize = 20)
plt.xlabel("Years", fontsize = 20)
plt.ylabel("Number of Attacks", fontsize = 20)
plt.show()
```

Total Terrorist attacks in each Region from 1970-2017

```python
reg = region.transpose()
reg["Total"] = reg.sum(axis = 1)
r = reg["Total"].sort_values(ascending = False)
r
```

```python
r.plot(kind = "bar", color = "orange", figsize = (20,10))
plt.title("Total number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

Total Casualities in each Region

```python
rcasual = terror[["Region","Casualty"]].groupby("Region").sum().sort_values(by = "Casualty", ascending = False)
rcasual
```

```python
rcasual.plot(kind = "bar", color = "yellow", figsize = (15,6))
plt.title("Region Wise Casualities", fontsize = 13)
plt.xlabel("Regions", fontsize = 13)
plt.xticks(fontsize = 12)
plt.ylabel("No.of Casualities", fontsize = 13)
plt.show()
```

Killed in each Region

```python
rkill = terror[["Region","Killed"]].groupby("Region").sum().sort_values(by = "Killed", ascending = False)
rkill
```

Wounded in each Region

```python
rwound = terror[["Region","Wounded"]].groupby("Region").sum().sort_values(by = "Wounded", ascending = False)
rwound
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

# Killed
rkill.plot(kind = 'bar', color = 'yellow', figsize = (15,6), ax = ax0)
ax0.set_title("People killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("No.of People Killed")

# Wounded
rwound.plot(kind = 'bar', color = 'orange', figsize = (15,6), ax = ax1)
ax1.set_title("People wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("No.of People Wounded")

plt.show()
```
